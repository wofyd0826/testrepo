### Import numpy and pandas

```python
import numpy as np
import pandas as pd
```

### Download the dataframe

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
```

### Remove the 'Not assigned' rows

```python
df[0].replace('Not assigned', np.nan, inplace = True)
df1 = df[0].dropna()
df1.head()
```

```python
df1.shape
```

### Merge two dataframe

```python
df2 = pd.read_csv('http://cocl.us/Geospatial_data')
df3 = pd.merge(df1, df2, on = 'Postal Code')
df3.head()
```

### Create a map

```python
import json

!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim

import requests
from pandas.io.json import json_normalize

import matplotlib.cm as cm
import matplotlib.colors as colors

from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes
import folium

print('Libraries imported.')
```

```python
map_toronto = folium.Map(location=[43.7, -79.4], zoom_start=10)

for lat, lng, borough, neighborhood in zip(df3['Latitude'], df3['Longitude'], df3['Borough'], df3['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
